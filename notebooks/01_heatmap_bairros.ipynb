```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Carregar o CSV
df = pd.read_csv("../data/imoveis_lisboa_pronto.csv")

# 2. Contar nmero de imveis por bairro
contagem = df['neighborhood'].value_counts().reset_index()
contagem.columns = ['neighborhood', 'num_imoveis']

# 3. Ordenar por quantidade
contagem = contagem.sort_values(by='num_imoveis', ascending=False)

# 4. Criar heatmap (1D visual)
plt.figure(figsize=(8,12))
sns.heatmap(
    contagem[['num_imoveis']].set_index(contagem['neighborhood']),
    cmap="RdYlGn", linewidths=0.5, annot=True, fmt="d", cbar=False
)
plt.title("Nmero de imveis por bairro (quanto mais verde, mais anncios)")
plt.xlabel("Quantidade")
plt.ylabel("Bairro")
plt.tight_layout()

# 5. Salvar
plt.savefig("../reports/heatmap_bairros.png")
plt.show()
```

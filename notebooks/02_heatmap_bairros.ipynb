```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carregar os dados
df = pd.read_csv("../data/imoveis_lisboa_pronto.csv")

# 2. Agrupar por bairro e tipo de imóvel
grouped = df.groupby(['neighborhood', 'property_type']).size().unstack(fill_value=0)

# 3. Ordenar os bairros por total de imóveis (soma das colunas)
grouped = grouped.loc[grouped.sum(axis=1).sort_values(ascending=False).index]

# 4. Criar gráfico de barras empilhadas horizontal
plt.figure(figsize=(12, 10))
grouped.plot(kind='barh', stacked=True, colormap='viridis', figsize=(10,12))

plt.title("Distribuição de imóveis por tipo em cada bairro")
plt.xlabel("Número de imóveis")
plt.ylabel("Bairro")
plt.legend(title="Tipo de imóvel")
plt.tight_layout()

# 5. Salvar
plt.savefig("../reports/imoveis_por_tipo_por_bairro.png")
plt.show()
```

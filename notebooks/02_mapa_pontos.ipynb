```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# 1) Carregar dados
df = pd.read_csv("../data/locations.csv")

# 2) Criar mapa (centro provisório)
m = folium.Map(location=[38.72, -9.14], zoom_start=12)

# 3) Adicionar marcadores e coletar para heatmap
heat_data = []
for _, row in df.iterrows():
    lat, lon, val = row.latitude, row.longitude, row.valor_medio
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        fill=True,
        fill_opacity=0.7,
        popup=f"{row.name}: €{val}"
    ).add_to(m)
    heat_data.append([lat, lon, val])

# 4) Adicionar HeatMap
HeatMap(heat_data, radius=25).add_to(m)

# 5) Ajustar bounds para todos os pontos
min_lat, max_lat = df.latitude.min(), df.latitude.max()
min_lon, max_lon = df.longitude.min(), df.longitude.max()
# fit_bounds espera [[south, west], [north, east]]
m.fit_bounds([[min_lat, min_lon], [max_lat, max_lon]])

# 6) Exibir
m
# 7) Salvar mapa
#m.save("../reports/lisbon_rent_map.html")

```
